```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = "data/problem3data.txt"
train_data = genfromtxt(train_file, delimiter='\t')

Y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
# Lasso regularization

regularization_strengths = np.linspace(0.01, 3, 30)
weights = []
for i in range(len(regularization_strengths)):
    a = regularization_strengths[i]
    clf = Lasso(alpha=a)
    clf.fit(X_train, Y_train)
    w = clf.coef_
    weights.append(w)

weights = np.asarray(weights).transpose()
for j in range(len(weights)):
    plt.plot(regularization_strengths, weights[j])
    plt.title("Lasso Regularization")
    plt.xlabel("regularization strength")
    plt.ylabel("weights")
```

```python
# Ridge regression

regularization_strengths = np.arange(1, 1e4)

weights = []
for i in range(len(regularization_strengths)):
    a = regularization_strengths[i]
    clf = Ridge(alpha=a)
    clf.fit(X_train, Y_train)
    w = clf.coef_
    #print(w)
    weights.append(w)

weights = np.asarray(weights).transpose()
for j in range(len(weights)):
    plt.plot(regularization_strengths, weights[j])
    plt.title("Ridge Regularization")
    plt.xlabel("regularization strength")
    plt.ylabel("weights")
```
